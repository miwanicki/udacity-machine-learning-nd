```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
periods = 365
time_series = 1

df = pd.DataFrame(
    data=np.random.randn(periods, time_series),
    columns=pd.Series(['s' + str(i) for i in range(time_series)], name="walk"),
    index=pd.date_range("2017-01-01", periods=periods,
                        name="date"))
```

```python
df = df.cumsum(axis=0).stack().reset_index(name="val")
```

```python
def dateplot(x, y, **kwargs):
    ax = plt.gca()
    data = kwargs.pop("data")
    data.plot(x=x, y=y, ax=ax, grid=False, **kwargs)
```

```python
g = sns.FacetGrid(df, hue="walk", size=10)
g = g.map_dataframe(dateplot, "date", "val")
```
